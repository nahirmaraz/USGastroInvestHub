# 2022

```python
import pandas as pd
from ntscraper import Nitter
```

```python
scraper= Nitter()
```

```python
terms=['restaurants','ethnic']
tw_22=scraper.get_tweets(terms, mode='term', since='2022-01-01', until='2022-12-31', near='USA')
```

```python
tw_22
```

```python
tweets_22 = []
for tweet_info in tw_22[1]['tweets']:
    text = tweet_info['text']
    date = tweet_info['date']
    likes = tweet_info['stats']['likes']
    comments = tweet_info['stats']['comments']
    retweets = tweet_info['stats']['retweets']
    data = [text, date, likes, comments, retweets]
    tweets_22.append(data)
```

```python
data= pd.DataFrame(tweets_22, columns= ['text','date','Nlikes','Ncomments','Nretweets'])
data
```

```python
data.info()
```

```python
data['date'].min()
```

```python
data['date'].max()
```

```python
data.to_csv('2022.csv')
```
