```python
import pandas as pd
import numpy as np
```

```python
metro_df = pd.read_csv('datasets/raw/DP03_All_Metros.csv')
metro_df.info()
```

```python
metro_df.replace(-999999999.0, np.nan, inplace = True)
```

```python
for column in metro_df.columns[4:]:
    metro_df[column].replace(np.nan, metro_df[column].mean(), inplace = True)
```

```python
lat_long = pd.read_csv('datasets/procesados/lat_long.csv', header = None)
lat_long.head()
```

```python
metro_df.insert(5, 'lat_long', lat_long)
metro_df.head()
```

```python
def lat_long_str_to_list_of_floats(lat_long_str):
    
    lat_long_str = lat_long_str.replace('[', '').replace(']', '')    
    lat_long_list = lat_long_str.split(',')
    
    lat_long_pairs_count =  int(len(lat_long_list) / 2)    
    lat_long_pairs = [[] for i in range(lat_long_pairs_count)]
    
    count = 0
    for pair in lat_long_pairs:
        
        pair.append(float(lat_long_list[0 + count]))
        pair.append(float(lat_long_list[1 + count]))
        count += 2
        
    return lat_long_pairs
```

```python
metro_df['lat_long'] = pd.Series(map(lat_long_str_to_list_of_floats, metro_df['lat_long']))
metro_df.head()
```

```python
FL_metro_df = metro_df[(metro_df['state'] == 'FL') & (metro_df['metro-micro'] == 'Metro')]
FL_metro_df.reset_index(drop = True, inplace = True)
FL_metro_df.head()
```

```python
metadata_restaurantes = pd.read_parquet('datasets/procesados/metadata-restaurants-FL.parquet')
```

```python
metadata_restaurantes.info()
```

```python
# 1. Iterar sobre las filas del DataFrame FL_metro_df
for index, row in FL_metro_df.iterrows():
    # 2. Separar los nombres de ubicación usando el guion "-" como separador
    location_name = row['geographic area name']
    location_parts = location_name.split('-')
    
    # 3. Inicializar la variable para contar las coincidencias
    total_matches = 0
    
    # 4. Para cada parte del nombre de la ubicación, buscar coincidencias en la columna "city" del DataFrame metadata_restaurantes
    for part in location_parts:
        total_matches += metadata_restaurantes['city'].str.count(part).sum()
    
    # 5. Calcular el cociente entre la columna population_16_up y el total de coincidencias
    pop_revenue_ratio = row['population_16_up'] / total_matches if total_matches != 0 else 0
    
    # 6. Agregar el resultado a una nueva columna llamada "pop_revenue_ratio" en el DataFrame FL_metro_df
    FL_metro_df.at[index, 'pop_venue_ratio'] = pop_revenue_ratio
```

```python
# Crear una lista para almacenar los datos expandidos
expanded_data = []

# Iterar sobre cada fila del DataFrame original
for index, row in FL_metro_df.iterrows():
    # Obtener el nombre de la ubicación y las coordenadas latitud/longitud
    location_name = row['geographic area name']
    lat_longs = row['lat_long']
    pop_revenue_ratio = row['pop_venue_ratio']  # Agregar la columna pop_revenue_ratio
    
    # Dividir el nombre de la ubicación en varias partes usando el guion como delimitador
    location_parts = location_name.split('-')
    
    # Iterar sobre cada par latitud/longitud y las partes del nombre de la ubicación
    for lat_long, location_part in zip(lat_longs, location_parts):
        # Crear una nueva fila con el nombre de la ubicación, las coordenadas y pop_revenue_ratio
        expanded_data.append([location_part.strip(), lat_long[0], lat_long[1], pop_revenue_ratio])

# Crear un nuevo DataFrame con los datos expandidos
FL_metro_df_expanded = pd.DataFrame(expanded_data, columns=['city', 'latitude', 'longitude', 'pop_venue_ratio'])
```
